```python
import re
import pandas as pd
from bnlp import BengaliCorpus as corpus

# Define the CleanText class
class CleanText:
    def __init__(self):
        punc = corpus.punctuations + ("‘") + ("’")
        self.PUNCTUATIONS = set(punc)
        self.STOPWORDS = set(corpus.stopwords)

    def remove_digits(self, text):
        return re.sub(r'[০-৯]+\d+', '', text).strip()

    def remove_punctuations(self, text, replace_with=" "):
        for punc in self.PUNCTUATIONS:
            text = text.replace(punc, replace_with)
        return ' '.join(text.split())

    def remove_stopwords(self, text):
        words = text.split()
        new_text = [word for word in words if word.lower() not in self.STOPWORDS]
        return ' '.join(new_text)

    def remove_english_and_special_chars(self, text):
        return re.sub(r'[a-zA-Z0-9!@#$%^&*()_+{}\[\]:;"\'<>,.?~\\/-]+', '', text)
```

# Clean dataset

### Testing effects of Cleaning

```python
# Create a sample DataFrame
data = {'text_column': [
    "সঠিক তদন্ত করতে হবে। This is an English sentence.",
    "বিচারের আওতায় আনতে হবে যে এই কাজ টা করেছে।",
    "এটি একটি নম্বর: ১২৩৪",
    "এবং এটি একটি সংখ্যার উদাহরণ: ২৩৪৫৬৭৮৯০"
]}

df = pd.DataFrame(data)

# Create an instance of CleanText
ct = CleanText()

# Apply cleaning methods to 'text_column'
df['text_column'] = df['text_column'].apply(ct.remove_punctuations).apply(ct.remove_digits).apply(ct.remove_english_and_special_chars)
# df['text_column'] = df['text_column'].apply(ct.remove_stopwords)
# Display the updated DataFrame
print(df)
```

```python
train_df = pd.read_csv(train_dataset_location)
validation_df = pd.read_csv(validation_dataset_location)
test_df = pd.read_csv(test_dataset_location)
```

```python
def df_cleaner(df, column_name="sentence1"):
    df[column_name] = df[column_name].apply(ct.remove_punctuations)
    # df['text_column'] = df['text_column'].apply(ct.remove_stopwords)
    df[column_name] = df[column_name].apply(ct.remove_digits)
    df[column_name] = df[column_name].apply(ct.remove_english_and_special_chars)
    return df
```

```python
train_df = df_cleaner(train_df, "Review")
validation_df = df_cleaner(validation_df, "Review")
test_df = df_cleaner(test_df, "Review")
```
