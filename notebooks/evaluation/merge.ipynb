# Balanced Merging

```python
import pandas as pd

# Assuming df is your DataFrame with columns 'sentence1' and 'label'
# Count the number of sentences for each label
label_counts = train_100['label'].value_counts()

# label_counts now holds the number of sentences for each label
print(label_counts)
```

```python
one_need = max(5152, 4556, 2874) - 5152
two_need = max(5152, 4556, 2874) - 4556
zero_need = max(5152, 4556, 2874) - 2874

# Define the number of samples you want for each label
samples_per_label = {
    0: zero_need,  # replace x, y, z with the actual numbers
    1: one_need,
    2: two_need
}

samples_per_label
```

```python
# Sample the rows from train_100_pp
sampled_dfs = []
for label, n_samples in samples_per_label.items():
    # Filter the DataFrame for the label and sample n_samples from it
    sampled = train_100_pp[train_100_pp['label'] == label].sample(n=n_samples, random_state=1)
    sampled_dfs.append(sampled)

# Concatenate the sampled DataFrames
sampled_df = pd.concat(sampled_dfs)

# Concatenate the sampled_df with train_100
merged_df_balanaced = pd.concat([train_100, sampled_df], join='inner', sort=False)

# Shuffle the merged DataFrame and reset the index
shuffled_merged_df_balanaced = merged_df_balanaced.sample(frac=1, random_state=1).reset_index(drop=True)

# shuffled_merged_df is your final DataFrame
shuffled_merged_df_balanaced
```
